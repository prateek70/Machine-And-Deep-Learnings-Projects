# Plot the following Continuous Distributions

***1. Uniform(-4,10)***

```python
import numpy as np 
from scipy.stats import uniform
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,1, 100)
ax.plot(x, uniform.pdf(x,-4,10),'r-', lw=5, alpha=0.6, label='uniform pdf')
r = uniform.rvs(size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.title("Uniform(-4,10")
plt.show()
```

***2. Normal(5,2)***

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
x = np.linspace(-2,12.5,100)
plt.plot(x,norm.pdf(x,loc=5,scale=2),label="mu=5, sigma=2")
plt.title("Normal(5,2)")
plt.show()
```

***3. Exponential(0.5)***

```python
from scipy.stats import expon
import matplotlib.pyplot as plt
x = np.linspace(0.7,5,100)
plt.plot(x,expon.pdf(x,0.5))
plt.title("Exponential(0.5)")
plt.show()
```

***4. T(1)***

```python
from scipy.stats import t
import matplotlib.pyplot as plt
x = np.linspace(-4,4,100)
plt.plot(x,t.pdf(x,1))
plt.title("T(1)")
plt.show()
```

***5. Beta(1,2.5)***

```python
from scipy.stats import beta
import matplotlib.pyplot as plt
x = np.linspace(0,5,100)
plt.plot(x,beta.pdf(x, 1, 2.5))
plt.title("Beta(1,2.5)")
plt.show()
```

***6. Gamma(1,5)***

```python
from scipy.stats import gamma
import matplotlib.pyplot as plt
x = np.linspace(0.5,10,100)
plt.plot(x,gamma.pdf(x, 1, 1/5))
plt.title("Gamma(1,5)")
plt.show()
```

***7. X^2(2)***

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
plt.plot(x,chi2.pdf(x, 2))
plt.title("X^2(2)")
plt.show()
```
