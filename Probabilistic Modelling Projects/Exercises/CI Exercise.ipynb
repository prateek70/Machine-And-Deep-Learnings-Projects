# Step 1: Load the data into numpy ndarray

```python
import numpy as np
df = np.array([19.8, 10.1, 15.4, 15.4, 18.5, 14.9, 7.5, 7.9, 12.7, 11.9, 15.4, 11.4, 14.1, 11.4, 17.6, 16.7, 15.8, 19.5, 8.8, 13.6, 11.9, 11.4])
```

# Step 2: Calculate mean and standard deviation (degree of freedom n-1)

```python
n = len(df)
mean = np.mean(df)
std = np.std(df, ddof=1)
print("n = ",n)
print("Mean = ",mean)
print("Standard Deviation",std)
```

- Mean is 13.713 and Standard deviation is 3.553 for n-1 degree of freedom

# Step 3: Check whether the data is normally distributed. We can check using box plot and normal probability plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x=df)
plt.show()
print("The boxplot shows us that there are no outliners present in data. THe median is approx. in the middle of IQR. The Data is slightly left skewed.")
```

```python
from scipy import stats
stats.probplot(df, plot=plt)
```

- From both Box and Prob plot we can tell that the population set is normally distributed

# Step 4: Check the values of n>10K3^2 and n>10K4

```python
print("Kurtosis Value = ",10*np.square(stats.kurtosis(df)))
print("Skew Value = ", 10*np.absolute(stats.skew(df)))
```
